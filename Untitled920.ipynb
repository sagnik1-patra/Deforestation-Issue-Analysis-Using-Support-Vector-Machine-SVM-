```python
# ============================================================
# Deforestation Issue Analysis Using SVM - FULL FIXED CODE
# ============================================================

import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.inspection import permutation_importance
from sklearn.decomposition import PCA

warnings.filterwarnings("ignore")

# ============================================================
# Paths
# ============================================================

DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Deforestation Issue Analysis Using Support Vector Machine (SVM)\deforestation_dataset.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Deforestation Issue Analysis Using Support Vector Machine (SVM)"
PREFIX = "svm_"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ============================================================
# Load Dataset
# ============================================================

df = pd.read_csv(DATASET_PATH)

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
display(df.head())

# ============================================================
# Missing Values and Summary
# ============================================================

df.isnull().sum().to_csv(os.path.join(OUTPUT_DIR, PREFIX + "missing_values.csv"))
df.describe(include="all").to_csv(os.path.join(OUTPUT_DIR, PREFIX + "dataset_summary.csv"))

df = df.drop_duplicates()

# ============================================================
# Select Target
# ============================================================

if "Forest_Loss_Area_km2" in df.columns:
    TARGET = "Forest_Loss_Area_km2"
elif "Tree_Cover_Loss_percent" in df.columns:
    TARGET = "Tree_Cover_Loss_percent"
else:
    raise ValueError("Target not found. Expected Forest_Loss_Area_km2 or Tree_Cover_Loss_percent.")

print("Selected Target:", TARGET)

# ============================================================
# Split Features and Target
# ============================================================

X = df.drop(columns=[TARGET])
y = df[TARGET]

categorical_cols = X.select_dtypes(include=["object"]).columns.tolist()
numerical_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()

print("Categorical Columns:", categorical_cols)
print("Numerical Columns:", numerical_cols)

# ============================================================
# Handle Missing Values
# ============================================================

X_encoded = X.copy()

for col in numerical_cols:
    X_encoded[col] = X_encoded[col].fillna(X_encoded[col].mean())

label_encoders = {}

for col in categorical_cols:
    X_encoded[col] = X_encoded[col].fillna(X_encoded[col].mode()[0])
    X_encoded[col] = X_encoded[col].astype(str)

    le = LabelEncoder()
    X_encoded[col] = le.fit_transform(X_encoded[col])
    label_encoders[col] = le

joblib.dump(label_encoders, os.path.join(OUTPUT_DIR, PREFIX + "label_encoders.pkl"))

cleaned_df = X_encoded.copy()
cleaned_df[TARGET] = y
cleaned_df.to_csv(os.path.join(OUTPUT_DIR, PREFIX + "cleaned_dataset.csv"), index=False)

# ============================================================
# Train Test Split
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.20,
    random_state=42,
    shuffle=True
)

# ============================================================
# Evaluation Function
# ============================================================

def evaluate_model(model_name, y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)

    return {
        "Model": model_name,
        "MAE": mae,
        "MSE": mse,
        "RMSE": rmse,
        "R2_Score": r2
    }

# ============================================================
# Linear SVM Model
# ============================================================

linear_model = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVR(kernel="linear"))
])

linear_model.fit(X_train, y_train)

linear_train_pred = linear_model.predict(X_train)
linear_test_pred = linear_model.predict(X_test)

linear_train_metrics = evaluate_model("Linear SVM Train", y_train, linear_train_pred)
linear_test_metrics = evaluate_model("Linear SVM Test", y_test, linear_test_pred)

# ============================================================
# Hyperparameter Tuning
# ============================================================

svm_model = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVR())
])

param_grid = {
    "svm__kernel": ["linear", "rbf", "poly"],
    "svm__C": [0.1, 1, 10, 100],
    "svm__gamma": ["scale", "auto", 0.01, 0.1],
    "svm__degree": [2, 3]
}

grid_search = GridSearchCV(
    svm_model,
    param_grid,
    cv=5,
    scoring="r2",
    n_jobs=-1,
    verbose=1
)

grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_

grid_results = pd.DataFrame(grid_search.cv_results_)
grid_results.to_csv(os.path.join(OUTPUT_DIR, PREFIX + "grid_search_results.csv"), index=False)

print("Best Parameters:", grid_search.best_params_)
print("Best CV Score:", grid_search.best_score_)

# ============================================================
# Best Model Evaluation
# ============================================================

best_train_pred = best_model.predict(X_train)
best_test_pred = best_model.predict(X_test)

best_train_metrics = evaluate_model("Best SVM Train", y_train, best_train_pred)
best_test_metrics = evaluate_model("Best SVM Test", y_test, best_test_pred)

metrics_df = pd.DataFrame([
    linear_train_metrics,
    linear_test_metrics,
    best_train_metrics,
    best_test_metrics
])

metrics_df.to_csv(os.path.join(OUTPUT_DIR, PREFIX + "model_metrics.csv"), index=False)

print("Model Metrics:")
display(metrics_df)

# ============================================================
# Cross Validation
# ============================================================

cv_scores = cross_val_score(
    best_model,
    X_encoded,
    y,
    cv=5,
    scoring="r2"
)

cv_df = pd.DataFrame({
    "Fold": list(range(1, len(cv_scores) + 1)),
    "R2_Score": cv_scores
})

cv_df.to_csv(os.path.join(OUTPUT_DIR, PREFIX + "cross_validation_scores.csv"), index=False)

# ============================================================
# Predictions
# ============================================================

prediction_df = X_test.copy()
prediction_df["Actual_" + TARGET] = y_test.values
prediction_df["Predicted_" + TARGET] = best_test_pred
prediction_df["Error"] = prediction_df["Actual_" + TARGET] - prediction_df["Predicted_" + TARGET]

prediction_df.to_csv(os.path.join(OUTPUT_DIR, PREFIX + "predictions.csv"), index=False)

# ============================================================
# Feature Importance
# ============================================================

perm = permutation_importance(
    best_model,
    X_test,
    y_test,
    n_repeats=20,
    random_state=42,
    scoring="r2"
)

feature_importance_df = pd.DataFrame({
    "Feature": X_encoded.columns,
    "Importance": perm.importances_mean
}).sort_values(by="Importance", ascending=False)

feature_importance_df.to_csv(os.path.join(OUTPUT_DIR, PREFIX + "feature_importance.csv"), index=False)

display(feature_importance_df)

# ============================================================
# Correlation Heatmap
# ============================================================

plt.figure(figsize=(12, 8))
corr = cleaned_df.corr(numeric_only=True)
plt.imshow(corr, aspect="auto")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + "correlation_heatmap.png"), dpi=300)
plt.show()

# ============================================================
# Feature Importance Graph
# ============================================================

plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df["Feature"], feature_importance_df["Importance"])
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.title("SVM Feature Importance")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + "feature_importance_graph.png"), dpi=300)
plt.show()

# ============================================================
# Actual vs Predicted Graph
# ============================================================

plt.figure(figsize=(8, 6))
plt.scatter(y_test, best_test_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted Deforestation")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + "actual_vs_predicted.png"), dpi=300)
plt.show()

# ============================================================
# Prediction Error Graph
# ============================================================

plt.figure(figsize=(10, 5))
plt.plot(prediction_df["Error"].values, marker="o")
plt.xlabel("Test Sample")
plt.ylabel("Error")
plt.title("Prediction Error Graph")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + "prediction_error_graph.png"), dpi=300)
plt.show()

# ============================================================
# Model Comparison Graph
# ============================================================

comparison_df = metrics_df[metrics_df["Model"].str.contains("Test")]

plt.figure(figsize=(8, 5))
plt.bar(comparison_df["Model"], comparison_df["R2_Score"])
plt.ylabel("R2 Score")
plt.title("SVM Model Comparison")
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + "model_comparison_graph.png"), dpi=300)
plt.show()

# ============================================================
# Top Feature Scatter Graphs
# ============================================================

top_features = feature_importance_df["Feature"].head(5).tolist()

for feature in top_features:
    plt.figure(figsize=(8, 5))
    plt.scatter(cleaned_df[feature], cleaned_df[TARGET])
    plt.xlabel(feature)
    plt.ylabel(TARGET)
    plt.title(feature + " vs " + TARGET)
    plt.tight_layout()
    safe_name = feature.replace("/", "_").replace("\\", "_").replace(" ", "_")
    plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + safe_name + "_vs_target.png"), dpi=300)
    plt.show()

# ============================================================
# SVM Decision Surface using PCA
# ============================================================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

pca_svm = SVR(
    kernel=grid_search.best_params_["svm__kernel"],
    C=grid_search.best_params_["svm__C"],
    gamma=grid_search.best_params_["svm__gamma"],
    degree=grid_search.best_params_["svm__degree"]
)

pca_svm.fit(X_pca, y)

x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 100),
    np.linspace(y_min, y_max, 100)
)

Z = pca_svm.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(9, 6))
plt.contourf(xx, yy, Z, alpha=0.6)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("SVM Decision Surface using PCA")
plt.colorbar(label=TARGET)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + "svm_decision_surface_pca.png"), dpi=300)
plt.show()

# ============================================================
# Save Models
# ============================================================

joblib.dump(best_model, os.path.join(OUTPUT_DIR, PREFIX + "best_svm_model.pkl"))
joblib.dump(linear_model, os.path.join(OUTPUT_DIR, PREFIX + "linear_svm_model.pkl"))

# ============================================================
# Save JSON Summary
# ============================================================

summary = {
    "project": "Deforestation Issue Analysis Using Support Vector Machine",
    "target_variable": TARGET,
    "dataset_shape": list(df.shape),
    "best_parameters": grid_search.best_params_,
    "best_cv_score": float(grid_search.best_score_),
    "cross_validation_mean_r2": float(cv_scores.mean()),
    "best_test_metrics": {
        "MAE": float(best_test_metrics["MAE"]),
        "MSE": float(best_test_metrics["MSE"]),
        "RMSE": float(best_test_metrics["RMSE"]),
        "R2_Score": float(best_test_metrics["R2_Score"])
    },
    "top_features": feature_importance_df.head(10).to_dict(orient="records")
}

with open(os.path.join(OUTPUT_DIR, PREFIX + "summary.json"), "w") as f:
    json.dump(summary, f, indent=4)

# ============================================================
# Save YAML Config
# ============================================================

config = {
    "project": "Deforestation Issue Analysis Using Support Vector Machine",
    "algorithm": "Support Vector Machine Regression",
    "target": TARGET,
    "train_test_split": "80-20",
    "features": X_encoded.columns.tolist(),
    "best_parameters": grid_search.best_params_
}

with open(os.path.join(OUTPUT_DIR, PREFIX + "config.yaml"), "w") as f:
    yaml.dump(config, f)

# ============================================================
# Final Report - SAFE VERSION
# ============================================================

top_feature_lines = []

for _, row in feature_importance_df.head(10).iterrows():
    top_feature_lines.append(
        "- " + str(row["Feature"]) + ": Importance Score = " + str(round(row["Importance"], 4))
    )

report_lines = [
    "# Deforestation Issue Analysis Using Support Vector Machine",
    "",
    "## Project Overview",
    "",
    "This project analyzes deforestation data using Support Vector Machine Regression.",
    "",
    "The dataset contains country-wise and year-wise deforestation-related features such as CO2 emissions, rainfall, population, GDP, policy strictness, corruption index, illegal lumbering incidents, and other environmental indicators.",
    "",
    "## Target Variable",
    "",
    "Selected target variable: " + TARGET,
    "",
    "## Data Preprocessing",
    "",
    "- Loaded the dataset",
    "- Checked missing values",
    "- Removed duplicate records",
    "- Encoded categorical columns using Label Encoding",
    "- Filled missing numerical values using mean imputation",
    "- Filled missing categorical values using mode imputation",
    "- Standardized features using StandardScaler",
    "- Split the dataset into 80% training and 20% testing data",
    "",
    "## Model Building",
    "",
    "Two SVM models were trained:",
    "",
    "1. Linear Support Vector Regression",
    "2. Tuned Support Vector Regression using GridSearchCV",
    "",
    "## Best Hyperparameters",
    "",
    str(grid_search.best_params_),
    "",
    "## Model Evaluation",
    "",
    "| Metric | Value |",
    "|---|---|",
    "| MAE | " + str(round(best_test_metrics["MAE"], 4)) + " |",
    "| MSE | " + str(round(best_test_metrics["MSE"], 4)) + " |",
    "| RMSE | " + str(round(best_test_metrics["RMSE"], 4)) + " |",
    "| R2 Score | " + str(round(best_test_metrics["R2_Score"], 4)) + " |",
    "",
    "## Cross Validation",
    "",
    "Average R2 Score: " + str(round(cv_scores.mean(), 4)),
    "",
    "## Top Important Features",
    "",
    "\n".join(top_feature_lines),
    "",
    "## Interpretation",
    "",
    "The model identifies the most influential factors affecting deforestation. Variables such as population, GDP, rainfall, policy strictness, corruption index, protected area percentage, and illegal lumbering incidents can strongly influence forest loss.",
    "",
    "Higher population may increase land demand for agriculture, housing, and infrastructure.",
    "",
    "GDP may influence deforestation through industrial expansion, urban development, and resource extraction. However, stronger economies may also invest more in forest conservation.",
    "",
    "Strict forest policies can help reduce illegal logging and uncontrolled land conversion.",
    "",
    "Corruption can weaken forest governance and increase illegal exploitation.",
    "",
    "Illegal lumbering incidents are direct indicators of forest damage and usually increase deforestation levels.",
    "",
    "Protected forest areas can reduce deforestation when conservation laws are properly implemented.",
    "",
    "## Recommendations",
    "",
    "1. Strengthen forest protection laws.",
    "2. Reduce illegal lumbering through monitoring and penalties.",
    "3. Increase protected forest area coverage.",
    "4. Improve transparency in forest governance.",
    "5. Promote sustainable agriculture.",
    "6. Use international aid for reforestation and forest monitoring.",
    "7. Balance economic growth with environmental protection.",
    "",
    "## Generated Files",
    "",
    "- svm_cleaned_dataset.csv",
    "- svm_missing_values.csv",
    "- svm_dataset_summary.csv",
    "- svm_model_metrics.csv",
    "- svm_grid_search_results.csv",
    "- svm_cross_validation_scores.csv",
    "- svm_predictions.csv",
    "- svm_feature_importance.csv",
    "- svm_best_svm_model.pkl",
    "- svm_linear_svm_model.pkl",
    "- svm_summary.json",
    "- svm_config.yaml",
    "- svm_correlation_heatmap.png",
    "- svm_feature_importance_graph.png",
    "- svm_actual_vs_predicted.png",
    "- svm_prediction_error_graph.png",
    "- svm_model_comparison_graph.png",
    "- svm_svm_decision_surface_pca.png"
]

report = "\n".join(report_lines)

with open(os.path.join(OUTPUT_DIR, PREFIX + "final_report.md"), "w", encoding="utf-8") as f:
    f.write(report)

# ============================================================
# Final Output
# ============================================================

print("\nAll results saved successfully in:")
print(OUTPUT_DIR)

print("\nGenerated Files:")
for file in os.listdir(OUTPUT_DIR):
    if file.startswith(PREFIX):
        print(file)
```
